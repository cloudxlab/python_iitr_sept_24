```python
def hcf(a, b):
    if a > b:
        a, b = b, a
    # a is smaller, b is bigger
    if a == 0:
        return 0
    remainder = b % a
    if remainder == 0:
        return a
    return hcf(remainder, a)
```

```python
def HCF(n, m):
    print(f"Calling {n}, {m}")
    if n > m:
        print("Swapping")
        n, m = m, n
    rem = m % n
    print(f"Remainder: {rem}")
    if rem == 0:
        return n
    else:
        return HCF(rem, n)
```

```python
def _factorial(n):
    
    if n < 0:
        print("Error!")
    elif n == 0:
        return 1
    elif k == n:
        return k
    return k * _factorial(n, k+1)
```

```python
def factorial(n):
    def _factorial(n1, k):
        if k == n1:
            return k
        return k * _factorial(n1, k+1)
    if n < 0:
        print("Error!")
    elif n == 0:
        return 1
    else:
        return _factorial(n, 1)
```
