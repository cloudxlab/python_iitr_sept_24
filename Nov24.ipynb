````python
class DB_Arraybased:
    def __init__(self):
        self.data = []
    def insert(self, val):
        self.data.append(val)
    def find(self, val):
        for v in self.data:
            if v == val:
                return True
        return False
````

````python
database = DB_Arraybased()
database.insert(10)
database.find(10) # TRUE
database.insert(20)
````

````python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
    def insert(self, node):
        if node.val < self.val:
            if self.left:
                self.left.insert(node)
            else:
                self.left = node
        else:
            if self.right:
                self.right.insert(node)
            else:
                self.right = node
    def find(self, val):
        print("..")
        if val == self.val:
            return self
        elif val < self.val:
            if self.left:
                return self.left.find(val)
            else:
                return None
        else:
            if self.right:
                return self.right.find(val)
            else:
                return None
    def show(self, tab=""):
        print(tab, self.val)
        if self.left:
            self.left.show("\t" + tab)
        if self.right:
            self.right.show("\t" + tab)
    def __repr__(self):
        return f"[{self.val}]"
````

````python
class BST:
    def __init__(self):
        self.root = None
    def insert(self, val):
        node = Node(val)
        if self.root:
            self.root.insert(node)
        else:
            self.root = node
````

````python
bst =  BST()
for i in [50, 10, 34, 56, -89, 100]:
    bst.insert(i)
    print("Inserted: ", i)
    bst.root.show()
````

````python
def my_sqrt(target):
    left = 0
    if target < 1:
        right = 1
    else:
        right = target 
    while True:
        guess = (left+right)/2

        result = guess*guess
#         print(f"left: {left}, Right: {right}, result: {result}, guess: {guess}")
        if abs(result - target) < 1e-8:
            return guess
        if result > target: # if our guess is bigger
            right = guess
        else:
            left = guess
````

````python
res = my_sqrt(200)
res * res
````
