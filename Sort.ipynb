````python
def sort(arr):
    i = 0
    while i < len(arr):
        j = i + 1
        while j < len(arr):
            if arr[j] < arr[i]:
                tmp = arr[i]
                arr[i] = arr[j]
                arr[j] = tmp
            j += 1
        i += 1
    return arr
````

````python
def sort(A):
    i, j = 0, len(A) - 1
    while i<j:
        if A[i] == 0:
            i += 1
            continue
        if A[j] == 1:
            j -= 1
            continue
        A[i], A[j] = A[j], A[i]
        i += 1
        j -= 1
    return A
````

````python
def merge(A, B): # Time O(n+m), Space O(n+m) 
    C = []
    i, j = 0, 0
    while i < len(A) and j < len(B):
        if A[i] < B[j]:
            C.append(A[i])
            i += 1
        else:
            C.append(B[j])
            j += 1
    R, r = None, 0
    if i < len(A):
        R = A
        r = i
    elif j < len(B):
        R = B
        r = j
    C.extend(R[r:])
#     while r < len(R):
#         C.append(R[r])
#         r += 1
    return C
````

````python
def merge_sort(A):
    print("Sort: ", A)
    if len(A) <= 1:
        return A
    mid = len(A)//2
    A1 = A[:mid]
    A2 = A[mid:]
    print(f"A1: {A1}", f"A2: {A2}")
    A1 = merge_sort(A1)
    A2 = merge_sort(A2)
    S = merge(A1, A2)
    print(f"Sorted: {S}")
    return S
````

````python
# Quick Sort
````

````python
If we have to sort a big file
- dictionary
- merge sort
    - split into two equal halves and sorted separately and merge them.
- quick sort
    - All words starting with k and above go to right and all words upto j to left
    - sort them individually
    - append them.
````

````python
def qs(A):
    if len(A) <= 1:
        return A
    p = A[0]
    mid = [p]
    left = []
    right = []
    for e in A[1:]:
        if e == p:
            mid.append(e)
        elif e < p:
            left.append(e)
        else:
            right.append(e)
    left = qs(left)
    right = qs(right)
    return left + mid + right
````

````python
qs([1,2,-3,2,1,- 3, 0])
````

````python
def qs1(A):
    if len(A) <= 1:
        return A
    p = A[0]
    left = []
    right = []
    for e in A[1:]:
        if e < p:
            left.append(e)
        else:
            right.append(e)
    left = qs1(left)
    right = qs1(right)
    return left + [p] + right
````

````python
class Tree:
    def __init__(self, arr):
        self.arr = arr
    def get_parent(self, i):
        idx = (i-1)//2
        return idx
    def get_children(self, i):
        idx = 2*i+1
        return idx, idx + 1
#         result = []
#         if idx <  len(self.arr):
#             result.append(self.arr[idx])
#             if idx + 1 < len(self.arr):
#                 result.append(self.arr[idx+1])
#             else:
#                 result.append(None)
#         else:
#             result.append([None, None])
#         return result
    def insert(self, num):
        arr = self.arr
        # Put the num at the end
        self.arr.append(num)
        idx = len(self.arr) - 1
        while True:
            pidx = self.get_parent(idx)
            if pidx < 0:
                break
            if arr[pidx] > arr[idx]:
                arr[idx], arr[pidx] = arr[pidx], arr[idx]
            else:
                break
            idx = pidx
            
    def elem(self, i):
        return self.arr[i]
````
